```python
#Draw a pie plot based on number of wins for each team

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df1 = pd.read_csv(r'E:\Match.csv',usecols=['Team1'])
teams=np.unique(df1.values) 

df5=pd.DataFrame()
df5['teams'] = teams

df2 = pd.read_csv(r'E:\Match.csv',usecols=['match_winner'])

a=[]
for j in teams:
    count = 0
    for i in range(len(df2)):
        if j == df2.values[i] :
            count= count+ 1
    a.append(count)
df5['wins'] = a
#df5

ds1=df5.teams
ds1
ds2=df5.wins
ds2

plt.pie(ds2,labels=ds1,shadow=True,startangle=90,autopct="%1.1f%%",labeldistance=1.2)
plt.legend(loc='best')
```
