```python
correlation_matrix = data1.corr().round(2)
# annot = True to print the values inside the square
sb.heatmap(data=correlation_matrix, annot=True)
```

### The correlation coefficient ranges from -1 to 1. 
### If the value is close to 1, it means that there is a strong positive correlation between the two variables.
### When it is close to -1, the variables have a strong negative correlation.

```python
plt.figure(figsize=(20, 5))

features = ['availability_365', 'number_of_reviews']
target = data1['reviews_per_month']

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = data1[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('reviews_per_month')
```

```python
X = pd.DataFrame(np.c_[data1['availability_365'], data1['number_of_reviews']], columns = ['availability_365','number_of_reviews'])
Y = data1['reviews_per_month']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```
