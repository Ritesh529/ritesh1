```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df1 = pd.read_csv(r'E:\AppleStore.csv',usecols=['price'])
df1
list1=[]
for i in df1.price :
    if i == 0.00 :
        list1.append(i)
    
#print(list1)     # ---------------free apps

list2=[]
for j in df1.price :
    if j != 0.00 :
        list2.append(j)
    
#print(list2)    # --------------- paid apps
```

```python
dd = len(df1)
legend = ['paid apps']
range = (0, dd) 
bins = 5
plt.hist([list2],bins,range, color=['green'], alpha=0.5)
plt.xlabel('price of  apps')
plt.ylabel('no.of paid apps')
plt.legend(legend)
plt.title('paid apps in the store')
plt.show()
```

```python
width = 0.30
plt.bar(0,list1,color='#1412cd',width = 0.30,label='free apps')
plt.bar(0,list2,color='red',width = 0.30,label='paid apps')
plt.legend()
```
