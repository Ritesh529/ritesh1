### <h3>Traditional Programming:</h3>  We feed in DATA (Input) + PROGRAM (logic), run it on machine and get output.

## There are two types of regression models:
###  <font color="dark grey">Linear Regression: When the output variable is continuous in nature, we use Linear Regression. </font>
###  <font color="dark grey">Logistics Regression: When the output variable is discrete in nature, we use Logistics Regression.</font>


### <h3>Linear Regression is further divided into two types:</h3>
### <font color="dark grey">1.Simple Linear Regression</font>
### <font color="dark grey">2.Multiple Linear Regression</font>


### <h3>Logistics Regression is further divided into two types:</h3>
### <font color="dark grey">1.Simple Logistics Regression</font>
### <font color="dark grey">2.Multiple Logistics Regression</font>


### <h2>Linear regression models are used to show or predict the relationship between two variables.</h2>
### *<font color="dark grey">the variable that are to predicted is called as dependent variables/ output variables.</font>
###  *<font color="dark grey">the variable that are used to predict the value of the dependent variable are called the independent</font>

### <h3>Simple Linear Regression:</h3>
### <font color="dark grey">If output is continuous and if we have only one input variable , then we go for SLR</font>
### <font color="dark grey">It measures the relationship between one input and one output variable.</font>


### Multiple Linear Regression :
### <font color="dark grey">If output is continuous and if we have more than one  input variable , then we go for MLR</font>


# <center><font color="light blue"><h1>Simple Linear Regression</h1></font></center>

### <ul>
### <font color="dark grey"><li>Simple Logistics Regression aims for statistical relationship between the variables , but not deterministic relationship .</li></font>
### <font color="dark grey"><li>Relationship between two variables is said to be deterministic if one variable can be accurately expressed by the other.</li> </font>
### <font color="dark grey"><li>Statistical relationship is not accurate in determining the relationship between two variables.</li></font>
### <font color="dark grey"><li>The best line can be said that has  less total errors as possible.</li></font>
### <font color="dark grey"><li>Error is the distance between the data points(.) to the Regression Line(straight line).</li></font>
### <font color="dark grey"><li>Whichever line has the minimum error, that will be chosen as the best fit line.</li></ul></font>

## <h2>Equation of SLR : y = b0 + b1x + E</h2>

### <ul>
### <li> y is the dependent variable</li>
### <li>x is the independent variable</li>
### <li>b0 indicates intercept</li>
### <li>b1 is the coefficient</li>
### <li>E(epsilon) is the error part.</li>
### </ul>


## Evaluating the SLR's performance:

### <ul><font color="dark grey"><li>R-squared value is used to measure the performance of the Linear Regression model.</li></font>
### <font color="dark grey"><li>It is also called as coefficient of determination.</li></font>
### <font color="dark grey"><li>R-squared measures the strength of the relationship between your model and the dependent variable on a convenient 0 – 100% scale.</li></font>
### <font color="dark grey"><li>R-squared value lies in between 0-100%. Higher the R-squared value, the better the model fits your data.</li></font>
### <font color="dark grey"><li>R-squared value >0.85 indicates it as a good model.</li></font>
### <font color="dark grey"><li>R-squared value lies between 0.65 and 0.8 indicates it as a moderate model.</li></font>
### <font color="dark grey"><li>R-squared value <0.65 indicates it as a weak model</li></font>
### <font color="dark grey"><li>But we cannot completely rely on this R-squared value.</li></font>
### </ul>
