```python
divisions=['A','B','C','D']
marks=[62,74,83,69]
avg_marks=[56,78,85,65]
index=np.arange(len(divisions))
width=0.70
plt.bar(divisions,marks,color="#1412cd")
plt.bar(divisions,avg_marks,color='g')
```

```python
index=np.arange(len(divisions))
width=0.3
plt.bar(divisions,marks,color="#1412cd",width=0.3,label='bar2')
plt.bar(index,avg_marks,color=["red","g"],width=0.3,label='bar1')
plt.legend()
```

```python
index=np.arange(len(divisions))
width=0.3
plt.barh(divisions,marks,color="#1412cd",width=0.3,label='bar2')
plt.barh(index+width,avg_marks,color=["red","g"],width=0.3,label='bar1')
plt.legend()
```
